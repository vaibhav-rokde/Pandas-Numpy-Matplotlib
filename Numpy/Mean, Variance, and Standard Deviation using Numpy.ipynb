```python
import numpy as np
```

```python
# Creating an array
data = np.array([3, 6, 9, 12, 15])
```

```python
# Mean
mean = np.mean(data) # Calculates the arithmetic mean (average) of the data.
print("Mean:", mean)
```

```python
# Variance
variance = np.var(data) # Computes the variance, which measures the spread or dispersion of the data.
print("Variance:", variance)
```

```python
# Standard Deviation
std_deviation = np.std(data) # Calculates the standard deviation, which is the square root of the variance and provides a measure of the data's dispersion around the mean.
print("Standard Deviation:", std_deviation)
```
