```python
df['ward_votes'] = df.groupby('WARD')['VOTING'].transform('sum')
df.head()
```

```python
df['Percentage_Votes'] = (df['VOTING'] / df['ward_votes']) * 100
df.head()
```

```python
df_sorted = df.sort_values(['WARD', 'VOTING'], ascending=[True, False])
df_sorted
```

```python
filtered_df = df_sorted.copy()
filtered_df
```

```python
filtered_df['Vote_Difference'] = filtered_df.groupby('WARD')['VOTING'].diff()
filtered_df['Vote_Difference'] = filtered_df['Vote_Difference'].fillna(0)
filtered_df
```

```python
filtered_df['Per_Votes_dif_f_winner'] = (filtered_df['Vote_Difference'] / filtered_df['ward_votes']) * 100
filtered_df.head()
```

```python
import numpy as np
```

```python
filtered_df['comapre_winning'] = np.where(filtered_df['Rank'] == 2, np.where(filtered_df['Per_Votes_dif_f_winner'] > -10, 1, 0), 0)
```

```python
# Create a dictionary mapping Ward to the candidate name
ward_candidate_dict = dict(zip(rank_1_2017['WARD'], rank_1_2017['CANIDATE NAME']))
```

```python

# Create a new column '2017_Rank1_Candidate' by mapping the ward to the candidate name
df['2017_Rank1_Candidate'] = df['WARD'].map(ward_candidate_dict)
```

```python

# Split the full name into first name and last name
df['First_Name'] = df['2017_Rank1_Candidate'].str.split().str[0]
df['First_Name']= df['First_Name'].astype(str)

df['First_Name']= df['First_Name'].apply(lambda x: x.capitalize())

df['Last_Name'] = df['2017_Rank1_Candidate'].str.split().str[-1]
df['Last_Name']= df['Last_Name'].astype(str)

df['Last_Name']= df['Last_Name'].apply(lambda x: x.capitalize())
# Combine first name and last name with a space separator

df['2017_Corporator'] = df['First_Name'] + ' ' + df['Last_Name']

# Drop the 'First_Name' and 'Last_Name' columns if needed
df = df.drop(['First_Name', 'Last_Name'], axis=1)
```
