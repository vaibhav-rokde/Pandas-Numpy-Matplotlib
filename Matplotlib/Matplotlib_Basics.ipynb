# Importing library

```python
import matplotlib.pyplot as plt
```

# Simple Plots

# Format Strings

Format strings in Matplotlib allow you to control the appearance of lines, markers, and colors in your plots. They are used as an optional argument in various plot functions to specify the style and format of the plotted elements. The format string consists of one or more characters, each representing a different attribute. Here are some commonly used format characters:

- **Line Style**:

    '-': Solid line 

    '--': Dashed line

    ':': Dotted line

    '-.': Dash-dot line

- **Marker Style**:

    '.': Point marker

    ',': Pixel marker

    'o': Circle marker

    's': Square marker

    '^': Triangle marker

    '+': Plus marker

    'x': Cross marker

- **Color**:

    'b': Blue

    'g': Green

    'r': Red

    'c': Cyan

    'm': Magenta

    'y': Yellow

    'k': Black

    'w': White

You can combine these characters to create a format string that suits your needs

```python

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, 'r--o')
plt.show()
```

```python
plt.plot(x, y, linestyle='--', marker='o', color='r')
```

# Label Parameters, Legend

In Matplotlib, you can add labels to your plots using various parameters and customize the legend to provide additional information about the plotted elements

```python
#Label Parameters
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#Title
plt.plot(x,y)
plt.title('My Plot')
```

```python
#Legend
plt.plot(x, y, label='Line')
plt.legend()
```

```python
plt.plot(x,y , label='line')
plt.legend(loc='upper right')
```

# Examples

```python
# Bar Chart
x = ['A', 'B', 'C', 'D']
y = [3, 7, 2, 5]

plt.bar(x, y)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

```python
# Pie Chart
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels)
plt.title('Pie Chart')
plt.show()
```

```python
# Histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 5]

plt.hist(data, bins=5)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# Scatter Plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
#donut chart
sizes = [30, 20, 15, 35]
labels = ['A', 'B', 'C', 'D']

fig, ax = plt.subplots()

# Outer ring (total)
wedges, outer_texts = ax.pie(sizes, labels=labels, wedgeprops=dict(width=0.4))

# Inner ring (inner circle)
inner_sizes = [sum(sizes)]  # Total size
inner_labels = ['Total']

inner_wedges, inner_texts = ax.pie(inner_sizes, radius=0.6, colors=['white'])

# Add a circle in the center (hole)
center_circle = plt.Circle((0, 0), 0.4, color='white')
ax.add_artist(center_circle)

# Set aspect ratio to be equal so that pie is drawn as a circle
ax.set_aspect('equal')

# Set title
plt.title('Donut Chart')

# Create legend
plt.legend(wedges, labels, loc='best')

plt.show()
```
